```python
from transformers import pipeline

# Initialize the TTS pipeline
tts = pipeline("text-to-speech")
```

```python
def text_to_speech(text, language='en'):
    # Check if the input is Arabic or English and select the model accordingly
    if language == 'ar':
        # For Arabic support
        tts = pipeline('text-to-speech', model="arabic-tts-model")
    else:
        # Default English TTS
        tts = pipeline('text-to-speech')

    # Generate speech from text
    audio = tts(text)

    # Save or return the audio
    with open('output_audio.wav', 'wb') as f:
        f.write(audio['audio'])

    return 'output_audio.wav'
```

```python
from gtts import gTTS

def text_to_speech_gtts(text, language='en'):
    tts = gTTS(text=text, lang=language)
    tts.save("output_audio.mp3")
    return "output_audio.mp3"
```

```python
# Import Gradio
import gradio as gr
import random

# To Define Language Dropdown with creative colors
language_options = [
    ("English", "🇬🇧"),  # Flag emoji for visual appeal
    ("Arabic", "🇸🇦")
]
language_selection = gr.Dropdown(
    choices=[lang[0] for lang in language_options],
    label='Select Language',
    type='value',
    value='English'  # Default value
)

# TO Define fun facts or quotes for each language
fun_facts = {
    "English": "Did you know? English is the most widely spoken language in the world!",
    "Arabic": "اللغة العربية تكتب من اليمين لليسار!"
}

# To Define the main function with dynamic output
def greet(language):
    greeting = "Hello!" if language == 'English' else "مرحبا!"
    fact = fun_facts[language]
    return f"{greeting}\n\n{fact}"

# Here we Create Gradio Interface
demo = gr.Interface(fn=greet, inputs=language_selection, outputs="text", title="Language Greeting App",
                    description="Select a language to receive a greeting and a fun fact!",
                    theme="default")  # Change theme if desired
demo.launch()
```

```python
import gradio as gr
from gtts import gTTS

def text_to_audio(text, language='en'):  # Set default language here
    # Generate audio
    tts = gTTS(text=text, lang=language, slow=False)
    audio_file = "output.mp3"
    tts.save(audio_file)

    # Return the audio file path
    return audio_file

# Set Up the Gradio Interface
language_options = ['en', 'ar']  # English and Arabic

# Create the Gradio interface
interface = gr.Interface(
    fn=text_to_audio,
    inputs=[
        gr.Textbox(label="Enter Text", placeholder="Type your message here..."),
        gr.Dropdown(label="Select Language", choices=language_options, type="value")
    ],
    outputs=gr.Audio(label="Generated Audio"),
    title="Text-to-Speech Converter",
    description="Convert your text to audio! Select the language and type in your text to hear it spoken.",
    theme="default"
)

# Launch the app
if __name__ == "__main__":
    interface.launch()
```

```python
import gradio as gr
from gtts import gTTS
import os

def text_to_audio(text, language, voice):
    # Choose the appropriate language
    if language == 'English':
        lang = 'en'
    elif language == 'Arabic':
        lang = 'ar'
    else:
        return None  # Handle unsupported language


    if voice == 'Female':
        tts = gTTS(text=text, lang=lang)
    elif voice == 'Male':
        tts = gTTS(text=text, lang=lang)


    # Save the generated audio to a file
    audio_file = "output.mp3"
    tts.save(audio_file)

    return audio_file
```

```python
# Voice options
voice_options = ['Male', 'Female']

# Here we Set up the Gradio interface
gr.Interface(
    fn=text_to_audio,
    inputs=[
        gr.Textbox(label="Enter Text", placeholder="Type your message here..."),
        gr.Dropdown(label="Select Language", choices=['English', 'Arabic'], type="value"),
        gr.Dropdown(label="Select Voice", choices=voice_options, type="value")
    ],
    outputs=gr.Audio(label="Generated Audio"),
    title="Text-to-Speech with Voice Options",
    description="Enter any text in English or Arabic, select the voice (Male/Female), and receive audio output."
).launch()
```
